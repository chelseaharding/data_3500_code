```python
import pandas as pd

df = pd.read_csv("/workspaces/data_3500_code/week10/content_videos/SLC_weather.csv")
df
```

```python
# longest hot streak
df["Hot"] = df["Max Temp"] > 65

current_streak = 0
longest_streak = 0

for is_hot in df["Hot"]:
    if is_hot:
        current_streak += 1
        if current_streak > longest_streak:
            longest_streak = current_streak
    else:
        current_streak = 0  # reset when itâ€™s not a hot day

print("Longest hot streak:", longest_streak, "days")
```

```python
# coldest night, warmest day
coldest_night = df.loc[df["Min Temp"].idxmin()]
warmest_day = df.loc[df["Max Temp"].idxmax()]

print("Coldest night:")
print(coldest_night[["Date", "Min Temp"]])
print("\nWarmest day:")
print(warmest_day[["Date", "Max Temp"]])
```

```python
# Calculate volatility for each day
df["Volatility"] = df["Max Temp"] - df["Min Temp"]

# Find the row with the highest volatility
most_volatile_day = df.loc[df["Volatility"].idxmax()]

print("Most volatile day:")
print(most_volatile_day[["Date", "Max Temp", "Min Temp", "Volatility"]])
```
